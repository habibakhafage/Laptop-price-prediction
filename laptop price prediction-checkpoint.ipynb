```python
def map_os(os):
    if 'Windows' in os:
        return 'Windows'
    elif 'mac' in os.lower():
        return 'macOS'
    elif 'linux' in os.lower():
        return 'Linux'
    elif 'android' in os.lower():
        return 'Android'
    elif 'chrome' in os.lower():
        return 'Chrome OS'
    else:
        return 'Other'  

df['OpSys'] = df['OpSys'].apply(map_os)
print(df['OpSys'].unique())
```

```python
df['ScreenResolution'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)')
df[['ScreenWidth', 'ScreenHeight']] = df['ScreenResolution'].str.split('x', expand=True)
df['ScreenWidth'] = df['ScreenWidth'].astype(int)
df['ScreenHeight'] = df['ScreenHeight'].astype(int)

df['screen_area'] = df['ScreenWidth'] * df['ScreenHeight']

df.drop(columns=['ScreenResolution'], inplace=True)
```

```python
def extract_Gpu_features(Gpu):
    brand = ''
    memory_size = 0
    
    if 'NVIDIA' in Gpu:
        brand = 'NVIDIA'
    elif 'AMD' in Gpu:
        brand = 'AMD'
    elif 'Intel' in Gpu:
        brand = 'Intel'
    
    memory_part = [s for s in Gpu.split() if 'GB' in s]
    if memory_part:
        memory_size = int(memory_part[0].replace('GB', '').strip())
    
    return pd.Series([brand, memory_size])

df[['GPU_brand', 'GPU_memory_size']] = df['Gpu'].apply(extract_Gpu_features)
```

# outliers removal

I first removed the outliers present in the target feature, Price Euros.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Price_euros'])
plt.title('Box Plot of Laptop Prices')
plt.show()
```

```python
Q1 = df['Price_euros'].quantile(0.25)
Q3 = df['Price_euros'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_mask = (df['Price_euros'] < lower_bound) | (df['Price_euros'] > upper_bound)

clean_data = df[~outliers_mask]
num_outliers_removed = outliers_mask.sum()

print(f'Number of outliers removed: {num_outliers_removed}')
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=clean_data['Price_euros'])
plt.title('Box Plot of Laptop Prices')
plt.show()
```

Then I considered removing outliers from the features most highly correlated with the target feature.

```python
df.corr()['Price_euros']
```

I selected CPU, RAM, screen width, screen height, screen area, and price per inch.

```python
features = ['Cpu', 'Ram', 'ScreenWidth', 'ScreenHeight', 'screen_area', 'Price_per_Inch']

plt.figure(figsize=(16, 12))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 3, i)  
    sns.boxplot(x=clean_data[feature])
    plt.title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    return df_cleaned

df_cleaned = remove_outliers(clean_data, 'Cpu')
df_cleaned = remove_outliers(df_cleaned, 'Ram')
df_cleaned = remove_outliers(df_cleaned, 'ScreenWidth')
df_cleaned = remove_outliers(df_cleaned, 'ScreenHeight')
df_cleaned = remove_outliers(df_cleaned, 'screen_area')
df_cleaned = remove_outliers(df_cleaned, 'Price_per_Inch')
```

# 7. neural network model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
import tensorflow as tf
```

```python
import tensorflow as tf
def r2_score(y_true, y_pred):
    total_sum_of_squares = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
    residual_sum_of_squares = tf.reduce_sum(tf.square(y_true - y_pred))
    return 1 - (residual_sum_of_squares / total_sum_of_squares)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model = keras.Sequential([
    layers.Input(shape=(170,)),  
    layers.Dense(32, activation='relu'),  
    layers.Dense(1)  
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=[r2_score])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
plt.plot(history.history.get('loss', []))
plt.plot(history.history.get('val_loss', []))
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.plot(history.history['r2_score'], label='Train R²')
plt.plot(history.history['val_r2_score'], label='Validation R²')
plt.title('R² Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('R² Score')
plt.legend()

plt.tight_layout()
plt.show()
```
